```python
import numpy as np

import matplotlib.pyplot as plot
```

```python
#Plot Sin
plot.subplot(2,2,1)
periode  = np.arange(0, 6, 0.01);

fsin   = np.sin(periode)

plot.plot(periode, fsin)

plot.title('Sine Function')


#Plot Cos
plot.subplot(2,2,2)
periode  = np.arange(0, 6, 0.01);

fcos   = np.cos(periode)

plot.plot(periode, fcos)

plot.title('Cos Function')


#Plot Tangent
plot.subplot(2,2,3)
periode  = np.arange(0, 6, 0.1);

ftan   = np.tan(periode)

plot.plot(periode, ftan)

plot.title('Tangent Function')

#Plot Tanh
plot.subplot(2,2,4)
periode  = np.arange(0, 6, 0.1);

ftanh   = np.tanh(periode)

plot.plot(periode, ftanh)

plot.title('Tanh Function')

plot.show()
```
