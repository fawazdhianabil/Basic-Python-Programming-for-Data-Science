```python
import pandas as pd
import plotly.express as px
```

```python
#Visualisasi Saham BBCA
#memili dataset
df_bbca = pd.read_csv('/content/drive/MyDrive/saham/BBCA.JK.csv')

#Memasukkan dataset dan 2 variabel
figure = px.scatter(df_bbca,
                    y = df_bbca['Open'],
                    x = df_bbca['Date'])
#Mengupdate fitur yang dibutuhkan
figure.update_layout(
    title='BBCA Stock Dataset', 
    title_font_color='#f3cec9',
    title_font_size=25,              
    xaxis=dict(
        title='Date',
        titlefont_size=16,
        tickfont_size=14),
    yaxis=dict(
        title='Open Price',
        titlefont_size=16,
        tickfont_size=14))

#Menampilkan plot
figure.show()
```
